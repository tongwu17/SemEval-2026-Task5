```python
import os
from datetime import datetime
import json
from llm_plausibility_scorer import compare_llm_models, LLM_MODEL_CONFIGS

OPENAI_API_KEY = ""
ANTHROPIC_API_KEY = ""
HF_TOKEN = ""

# Parameters
data_path = 'ambistory_data/test.json'
model_keys = ['gpt-4.1']
prompt_strategies = ['improved_criteria']
max_samples = None

results = compare_llm_models(
    data_path=data_path,
    model_keys=model_keys,
    prompt_strategies=prompt_strategies,
    max_samples=max_samples,
    openai_api_key=OPENAI_API_KEY,
    anthropic_api_key=ANTHROPIC_API_KEY,
    hf_token=HF_TOKEN
)


# Save results
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
model_str = "_".join(model_keys)
strategy_str = "_".join(prompt_strategies)
input_file_str = data_path.split("/")[-1].replace(".json", "")
filename = f"results_{input_file_str}_{model_str}_{strategy_str}_{max_samples}_{timestamp}.json"

with open(filename, 'w') as f:
    json.dump(results, f, indent=2)
    
print(f"\nResults saved to {filename}")
```
